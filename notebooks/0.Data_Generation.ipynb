```python
from pathlib import Path
import sys

# Add project src/ folder to Python path so imports work from notebooks/
PROJECT_ROOT = Path("..").resolve()
SRC_DIR = PROJECT_ROOT / "src"
if str(SRC_DIR) not in sys.path:
    sys.path.insert(0, str(SRC_DIR))
```

```python
"""Synthetic raw dataset generation + quick preview.

This notebook is ONLY for checking that the Faker-based
raw tables are generated correctly. All analytical steps
(EDA, cohorts, RFM, unit economics, churn) will be done
in separate notebooks, always reading from Supabase.
"""

from __future__ import annotations

from pathlib import Path

import pandas as pd

from fintech_ai_segmentation.faker_base_generation import generate_all_base_tables


# ===========================
# Generate all base tables
# ===========================

base_tables = generate_all_base_tables()

customers_raw = base_tables["customers_raw"]
transactions_raw = base_tables["transactions_raw"]
products_raw = base_tables["products_raw"]
customer_products_raw = base_tables["customer_products_raw"]


# ===========================
# Quick preview of all raw tables
# ===========================

print("\ncustomers_raw (first 5 rows):")
display(customers_raw.head())

print("\ntransactions_raw (first 5 rows):")
display(transactions_raw.head())

print("\nproducts_raw (first 5 rows):")
display(products_raw.head())

print("\ncustomer_products_raw (first 5 rows):")
display(customer_products_raw.head())
```

```python
# Save raw tables to CSV in data/raw
raw_data_dir = PROJECT_ROOT / "data" / "raw"
raw_data_dir.mkdir(parents=True, exist_ok=True)

customers_raw.to_csv(raw_data_dir / "customers_raw.csv", index=False)
transactions_raw.to_csv(raw_data_dir / "transactions_raw.csv", index=False)
products_raw.to_csv(raw_data_dir / "products_raw.csv", index=False)
customer_products_raw.to_csv(raw_data_dir / "customer_products_raw.csv", index=False)

print(f"Saved customers_raw to {raw_data_dir / 'customers_raw.csv'}")
print(f"Saved transactions_raw to {raw_data_dir / 'transactions_raw.csv'}")
print(f"Saved products_raw to {raw_data_dir / 'products_raw.csv'}")
print(f"Saved customer_products_raw to {raw_data_dir / 'customer_products_raw.csv'}")
```
